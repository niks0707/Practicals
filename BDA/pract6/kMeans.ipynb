## Generate some sample data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate some sample data f
n_samples = 300
n_clusters = 3
random_state = 42

X, _ = make_blobs(n_samples=n_samples, centers=n_clusters, random_state=random_state)
```

## Apply k-means clustering

```python
# Apply k-means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

## Define cluster names (customize these based on your data and insights)

```python
# Define cluster names (customize these based on your data and insights)
cluster_names = {
    0: "High-Spending Customers",
    1: "Frequent Shoppers",
    2: "Inactive Customers",
}
```

## Plot the data points and cluster centers with cluster names

```python
# Plot the data points and cluster centers with cluster names
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Cluster Centers')
plt.legend()

# Annotate cluster names
for i in range(n_clusters):
    plt.annotate(cluster_names[i], (kmeans.cluster_centers_[i, 0], kmeans.cluster_centers_[i, 1]), 
                 color='red', fontsize=12, weight='bold', ha='center', va='center')

plt.title('K-Means Clustering with Cluster Names')
plt.show()
```
